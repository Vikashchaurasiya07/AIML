```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Line graph

x = np.arange(1,11)

y = x * 2

plt.plot(x, y)

plt.xlabel('X - Axis')

plt.ylabel('Y - Axis')

plt.title(" Line Plot ")

plt.show()
```

```python
plt.plot(x, y, color='r', linewidth=1, linestyle="--")
```

```python
# for 2 Line graph

z = x*3

plt.plot(x, y, color='r')

plt.plot(x, z, color='b')

plt.show()
```

```python
#subplots
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot(x,z)
plt.show()
```

```python
#bar graph
names = ['Sachin', 'Dhoni', 'Virat', 'Yuvraj']
runs=[78,67,87,90]
plt.bar(names,runs)
plt.show()
```

```python
plt.barh(names,runs)
```

```python
# Scatter plot 
x = [10,20,30,40,50,60,70,80]
y=[8,2,5,6,8,9,4,2,]
plt.scatter(x,y)
plt.show()
```

```python
#Histogram
age=[1,3,3,3,9,0,5,4,4,8,8,8,12,12,12]
plt.hist(age,bins=20)
plt.show()
```

```python
# pie chat
company=['Apple', 'samsung', 'lg']
qty=[120,150,130]
plt.pie(qty,labels=company,autopct="%0.2f%%")
plt.show()
```

```python
# Box plot - outliers

x = [1, 2, 3, 4, 5, 6, 7, 8, 9]

y = [ 1, 2, 3, 4, 5, 4, 3, 2, 1]

z = [6, 7, 8, 9, 8, 7, 6, 5, 4]
data=list([x,y,z])
plt.boxplot(data)
plt.show()
```

## Line graph

```python
import matplotlib.pyplot as plt 

#Sample data 
months =["Jan", "Feb", "Mar", "April","May","June"]
energy_consumption = [1200,1300,1100,1500,1400,1600]

#plot aline plt
plt.plot(months, energy_consumption, marker='o', linestyle='--', color='b')
plt.xlabel('Months')
plt.ylabel('Energy Consumption(MWh)')
plt.title('Energy Consumption in 6 Months')
plt.show()
```

## Bar Chart


```python
#Sample data
energy_sources=['Solar','Wind','Hydropower','Biomass']
energy_values=[1200,3400,2900,2500]
#Creating bar chart
plt.bar(energy_sources,energy_values,color='green')
#add title axis labels
plt.title("Energy consumption by renewable energy sources")
plt.title("Energy source")
plt.title("Energy Consumption (MWh)")
plt.show()
```

```python
#pie chart to show energy consumption distribution
plt.pie(energy_values, labels=energy_sources, autopct='%1.1f%%',colors=['gold', 'lightblue', 'lightgreen', 'coral'])
plt.title('Energy consumption share by resource')
plt.show()
```

```python
# Sample data for energy consumption and carbon emissions 
carbon_emissions=[200,500,450,300] # in kg CO2
#create a scatter plot
plt.scatter(energy_values, carbon_emissions,color='b')
# add titles and labels
plt.title('Energy Consumption vs Carbon Emission')
plt.xlabel('Energy Consumption in MWh')
plt.ylabel('Carbon Emission in kg CO2')
plt.show()
```

```python
#create a bar chart for customization
plt.bar(energy_sources,energy_values,color='orange',edgecolor='black',linewidth=2)
#add titles and labels
plt.title('Customized energy consumption by source',fontsize=14,fontweight='bold')
plt.xlabel('Energy source',fontsize=12)
plt.ylabel('Energy Consumption in MWh',fontsize=12)
# Add grid and style
plt.grid(True,linestyle='--',alpha=0.7)
plt.legend(['Energy Consumption'],loc='upper left')
plt.show()
```
