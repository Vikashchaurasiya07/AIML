## Principal Componenet Analysis - PCA

```python
# Required Libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv('green transportation system.csv')
data.head()
```

```python
data.shape
```

```python
# Normalize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
print(pd.DataFrame(data_scaled, columns=data.columns).head())
```

```python
data_scaled
```

```python
# Initialize PCA
pca = PCA()
# Fit PCA on the sacled data
pca.fit(data_scaled)
# Explained variance ratio for ech principal component
print("Explained variance ratio for each principal component")
print(pca.explained_variance_ratio_)
# Cummulative explained variance
print('\n Cummulative explained ratio')
print(np.cumsum(pca.explained_variance_ratio_))
```

```python
#Plot the explained variance ratio
plt.figure(figsize=(8, 6))
plt.bar(range(1, len(pca.explained_variance_ratio_)+ 1),pca.explained_variance_ratio_)
plt.title('Explained Variance by each Principal Component')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.show()
```

```python
# Cumulative explained variance plot
plt.figure(figsize=(8,6))
plt.plot(range(1, len(pca.explained_variance_ratio_)+ 1),
         np.cumsum(pca.explained_variance_ratio_), marker='o',linestyle='--')
plt.title('Cumulative Explained Variance')
plt.xlabel('Number of PCs')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```

```python
# Project the data onto the finst two PCs
data_pca = pca.transform(data_scaled)
#Create a dataframe for first two PCs
pca_df = pd.DataFrame(data_pca[:, :2], columns=['PC1', 'PC2'])
# Visualize the data in @D space for 2 PCs
plt.figure(figsize=(8, 6))
plt.scatter(pca_df['PC1'],pca_df['PC2'], c='red', edgecolors='k', alpha=0.7)
plt.title('PCA of Green Transportation Syatem')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal component 2')
plt.show()
```
